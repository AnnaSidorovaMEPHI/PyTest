```python
class SocialNetworksConnector():
    def __init__(self, ID):
        self.id=ID
        self.vktoken=""
        self.fbtoken = ''        
        self.my_app_id = '388121675212208'
        self.my_app_secret = ''
        FacebookAdsApi.init(self.my_app_id, self.my_app_secret, self.fbtoken)#–°–æ–∑–¥–∞–Ω–∏–µ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –¥–ª—è —Ñ–µ–π—Å–±—É–∫–∞
        self.session = vk.Session(access_token=self.vktoken)#—Å–æ–∑–¥–∞–Ω–∏–µ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –¥–ª—è –≤–∫–æ–Ω—Ç–∞–∫—Ç–µ
        self.vk_api = vk.API(self.session)
        
    def get_profile(self,iD):
        try:
            data=self.vk_api.users.get(user_ids=iD,fields='status, online, sex, bdate, city, country',v=5.101)[0]
            try:
                status=data['status']
            except:
                status=''
            try:
                bdate=data['bdate']
            except:
                bdate=''
            try:
                city=data['city']['title']
            except:
                city=''
            try:
                country=data['country']['title']
            except:
                country=''
            print(data['first_name'],data['last_name'],'\n','–°—Ç–∞—Ç—É—Å:', status,'\n','–ü–æ–ª:',str(data['sex']).replace('1','–∂–µ–Ω.').replace('2','–º—É–∂.'),'\n','–î–∞—Ç–∞ —Ä–æ–∂–¥–µ–Ω–∏—è:',bdate,'\n', city, country,'\n\n')
        except:
            try:
                api_request="https://graph.facebook.com/"+str(iD)+"?fields=first_name,last_name&access_token="+self.fbtoken
                res = requests.get(api_request).text
                print((json.loads(res))["first_name"],(json.loads(res))["last_name"])
            except:
                print('–≠—Ç–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –Ω–µ—Ç –≤ –¥–æ—Å—Ç—É–ø–Ω—ã—Ö —Å–æ—Ü.—Å–µ—Ç—è—Ö')
    def GET_FRIENDS(self,n):
        print('–ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ –¥—Ä—É–∑–µ–π:')
        try:
            fr_ids=self.vk_api.friends.get(user_id=self.id,count=n,v=5.101)['items']
            for i in fr_ids:
                self.get_profile(i)
        except:
            api_request="https://graph.facebook.com/"+str(self.id)+"/friends?access_token="+self.fbtoken
            res = requests.get(api_request).text
            print((json.loads(res)))
    def GET_USER(self):
        self.get_profile(self.id)
    def GET_WALL(self):
        print('–ü–æ–ª—É—á–µ–Ω–∏–µ –∑–∞–ø–∏—Å–µ–π —Å–æ —Å—Ç–µ–Ω—ã:')
        try:
            wall=self.vk_api.wall.get(owner_id=self.id,v=5.101)
            return wall
        except:
            api_request="https://graph.facebook.com/"+str(self.id)+"/feed?access_token="+self.fbtoken
            res = requests.get(api_request).text
            print((json.loads(res)))
    def PARSE_WALL(self):
        wall=self.GET_WALL()
        for t in wall['items']:
            try:
                print(self.vk_api.users.get(user_ids=t['from_id'],v=5.101)[0]['first_name'],self.vk_api.users.get(user_ids=t['from_id'],v=5.101)[0]['last_name'],t['copy_history'][0]['text'])
            except:
                print(self.vk_api.users.get(user_ids=t['from_id'],v=5.101)[0]['first_name'],self.vk_api.users.get(user_ids=t['from_id'],v=5.101)[0]['last_name'],t['text'])#—Å–≤–æ–∏ –∑–∞–ø–∏—Å–∏

        
```
